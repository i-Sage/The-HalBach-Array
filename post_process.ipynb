```python
import matplotlib.pyplot as plt
import numpy as np


params = {
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "axes.grid": True,
    "grid.alpha": 0.3,
}
plt.rcParams.update(params)

def read_exported_data(path: str):
    data = np.loadtxt(path)
    
    if data.ndim == 1:
        data = data.reshape(1, -1)

    x = data[:, 0]
    y = data[:, 1]
    
    return x, y
```

```python
h_01x, h_01b = read_exported_data("exported_data/halbach_0.1mm_b.txt")
h_1x,  h_1b  = read_exported_data("exported_data/halbach_1mm_b.txt")
h_10x, h_10b = read_exported_data("exported_data/halbach_10mm_b.txt")

u_01x, u_01b = read_exported_data("exported_data/uniform_0.1mm_b.txt")
u_1x,  u_1b  = read_exported_data("exported_data/uniform_0.1mm_b.txt")
u_10x, u_10b = read_exported_data("exported_data/uniform_0.1mm_b.txt")


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,5))

# 0.1mm above array surface
ax1.plot(h_01x, h_01b, '-', lw=1, color='blue', label="HalBach Array")
ax1.plot(u_01x, u_01b, '--', lw=2, color='red', label="Uniform Array")
ax1.set_xlabel("Distance Along Array (mm)", fontsize=8)
ax1.set_ylabel("Flux Density |T|", fontsize=8)
ax1.legend(loc="upper right", fontsize=9)
ax1.set_title("0.01 mm Above Magnet Array", fontsize=15)

# 1mm above array surface
ax2.plot(h_1x, h_1b, '-', lw=1, color='blue', label="HalBach Array")
ax2.plot(u_1x, u_1b, '--', lw=2, color='red', label="Uniform Array")
ax2.set_xlabel("Distance Along Array (mm)", fontsize=8)
ax2.set_ylabel("Flux Density |T|", fontsize=8)
ax2.legend(loc="upper right", fontsize=9)
ax2.set_title("1 mm Above Magnet Array", fontsize=15)

# 10mm above array surface
ax3.plot(h_10x, h_10b, '-', lw=1, color='blue', label="HalBach Array")
ax3.plot(u_10x, u_10b, '--', lw=2, color='red', label="Uniform Array")
ax3.set_xlabel("Distance Along Array (mm)", fontsize=8)
ax3.set_ylabel("Flux Density |T|", fontsize=8)
ax3.legend(loc="upper right", fontsize=9)
ax3.set_title("10 mm Above Magnet Array", fontsize=15)
```
